```python
import sankey
import pandas as pd
```

```python
df = pd.read_csv(
    'fruits.txt', sep=' ', names=['true', 'predicted']
)
colorDict = {
    'apple':'#f71b1b',
    'blueberry':'#1b7ef7',
    'banana':'#f3f71b',
    'lime':'#12e23f',
    'orange':'#f78c1b',
    'kiwi':'#9BD937'
}

ax = sankey.sankey(
    df['true'],
    df['predicted'],
    colorDict=colorDict,
    color_gradient=True
)
```

```python
ax = sankey.sankey(
    df['true'],
    df['predicted'],
    colorDict=colorDict,
    color_gradient=False
)
```

```python
alphaDict = {
    'apple': 0.65,
    'blueberry': 0.65,
    'banana': 0,
    'lime': 1,
    'orange': 0.65,
    'kiwi': 0.65
}
```

```python
sankey.sankey(
    df['true'],
    df['predicted'],
    colorDict=colorDict,
    alphaDict=alphaDict,
    color_gradient=True,
)
```

```python
sankey.sankey(
    df['true'],
    df['predicted'],
    colorDict=colorDict,
    alphaDict=alphaDict,
)
```
